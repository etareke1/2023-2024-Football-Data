```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

### Importing the CSV file

```python
football_data_csv = pd.read_csv("datasets/big_5_players_stats_2023_2024.csv", encoding = "latin-1")
```

```python
football_data_csv.head(10)
```

### Removing useless columns

```python
football_data_csv.set_index("Player", inplace = True)
```

```python
football_data_csv.drop("Rank", axis = 1, inplace = True)
```

```python
football_data_csv.head()
```

### Questions to solve:
- Which players scored the most goals
- Which players have the most assists
- Does the number of minutes played correlate to number of goals and assists

### Which players have scored the most goals?

```python
top_goalscorers = football_data_csv.sort_values(by="Performance_Gls", ascending=False)
```

```python
top_goalscorers["Performance_Gls"].head(10)
```

```python
x = top_goalscorers.index[:10]
y = top_goalscorers["Performance_Gls"][:10]
plt.bar(x, y)
plt.xticks(rotation = 60)
plt.ylabel('Goals')
plt.title('Top 10 Goalscorers')
plt.show()
```

### Which players have the most assists?

```python
top_assists = football_data_csv.sort_values(by="Performance_Ast", ascending=False)
```

```python
top_assists["Performance_Ast"].head(10)
```

```python
x = top_assists.index[:10]
y = top_assists["Performance_Ast"][:10]
plt.bar(x, y)
plt.xticks(rotation = 60)
plt.ylabel('Assists')
plt.title('Top 10 Assists')
plt.show()
```

### Does minutes played correlate to total goals and assists?

#### Getting the data for minutes played

```python
minutes_played = football_data_csv["Playing Time_Min"]
minutes_played.shape
```

```python
minutes_played.head()
```

I concluded that defenders and goalkeepers should be removed from the data as those players are rarely involved in goals or assists

```python
fw_mf_players = football_data_csv[~football_data_csv["Position"].str.contains("DF|GK")]
fw_mf_players.head()
```

```python
fw_mf_minutes_played = football_data_csv[~football_data_csv["Position"].str.contains("DF|GK")]["Playing Time_Min"]
fw_mf_minutes_played.shape
```

```python
fw_mf_minutes_played.head()
```

#### Getting the data for Goals + Assists

```python
fw_mf_GA = fw_mf_players["Performance_G+A"]
```

```python
fw_mf_GA.head()
```

#### Joining the tables

```python
GA_vs_minutes = pd.merge(fw_mf_GA, fw_mf_minutes_played, on = "Player")
```

```python
GA_vs_minutes.head()
```

#### Finding the correlation coefficient

```python
GA_vs_minutes.corr()
```

This concludes that there is a positive correlation of 0.620287 between Goals + Assists and minutes played

#### Representing the data in a graph

```python
x = GA_vs_minutes["Playing Time_Min"]
y = GA_vs_minutes["Performance_G+A"]
plt.figure(figsize=(30, 15))
sns.regplot(x = x, y = y)
```
